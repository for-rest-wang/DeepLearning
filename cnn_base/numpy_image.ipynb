```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import sklearn
import sys
import tensorflow as tf
import time
from PIL import Image

from tensorflow import keras
from tensorflow.keras.preprocessing import image
print(tf.__version__)
```

# 最基础的numpy图像操作

```python
img = image.load_img("D:/image/yx.jpg")
im = np.asarray(img)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15,5)) 
for c, ax in zip(range(3), axs):
    one_array = im[:,:,c]
    pd.DataFrame(one_array).to_csv("D:/image/yx" + str(c) +".csv")
    saveim = Image.fromarray(one_array)
    saveim.save("D:/image/yx" + str(c) +".png")
    ax.imshow(one_array.astype(int))
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15,5)) 
for c, ax in zip(range(3), axs): # 使用zip来同时循环3通道和3个子图对象
    tmp_im = np.zeros(im.shape) # 初始化一个和原图像大小相同的三维数组
    # 注意 tmp_im 仍然是三通道
    tmp_im[:,:,c] = im[:,:,c] # 只复制某一通道
    one_channel = im[:,:,c].flatten() # 索引该通道并展平至一维
    print("channel", c, " max = ", max(one_channel), "min = ", min(one_channel)) # 输出该通道最大最小的像素值
    ax.imshow(tmp_im.astype(int)) # 在子图上绘制
    ax.set_axis_off() # 去掉子图坐标轴
# 注意以上 tmp_im 采用的是切片复制
```
