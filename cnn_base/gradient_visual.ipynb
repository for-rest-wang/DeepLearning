```python
import tensorflow as tf
import numpy as np

from math import  pi

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.ion()
```

```python
learning_rate = 0.01
x = tf.Variable(-10.)
X = []
Y = []
for i in range(200):
#     print("epoch:",i)
    #计算和更新梯度
    with tf.GradientTape() as tape:
        y = tf.square(x)
    x_grad = tape.gradient(y,x)
    x.assign_sub(x_grad * learning_rate)
    
    X.append(x.numpy())
    Y.append(y.numpy())
#     optimizer.apply_gradients(zip(gradients,[w,b]))     #更新梯度
x1 = tf.Variable(initial_value=np.arange(-10,10.))
y1 = tf.square(x1)
# print(X,Y)
plt.xlim(-10, 10)
plt.ylim(-20, 100)
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x1.numpy().flatten(),y1.numpy().flatten())
plt.plot(X,Y,'.')
plt.show()
```

```python
learning_rate = 0.05
x1 = tf.Variable(-4.)
y1 = tf.Variable(-4.)
X1 = []
Y1 = []
Z1 = []
for i in range(1000):
#     print("epoch:",i)
    #计算和更新梯度
    with tf.GradientTape() as tape:
        z1 = -tf.cos(((x1**2) + y1**2)**(1/6))
    x_grad,y_grad = tape.gradient(z1,[x1,y1])
    x1.assign_sub(x_grad * learning_rate)
    y1.assign_sub(y_grad * learning_rate)
    
    X1.append(x1.numpy())
    Y1.append(y1.numpy())
    Z1.append(z1.numpy())
```

```python
fig = plt.figure()
ax = Axes3D(fig)
x = np.arange(-10,10,0.25)
y = np.arange(-10,10,0.25)
X,Y = np.meshgrid(x,y)#创建网格，这个是关键
Z = -np.cos(((X**2) + Y**2)**(1/6))
plt.xlabel('x')
plt.ylabel('y')
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap='rainbow')
ax.scatter(X1,Y1,Z1,'.')
plt.show()
```
